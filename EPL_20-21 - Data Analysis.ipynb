```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
epl_data = pd.read_csv('EPL_20-21.csv')
```

```python
epl_data.head()
```

```python
epl_data.info()
```

```python
epl_data.describe()
```

```python
epl_data.isna().sum()
```

```python
# Creating 2 new columns

epl_data['MinPerMatch'] = (epl_data['Mins'] / epl_data['Matches']).astype(int)
epl_data['GoalsPerMatch'] = (epl_data['Goals'] / epl_data['Matches']).astype(float)
epl_data.head()
```

```python
# Total Goal

Total_Goals = epl_data['Goals'].sum()
Total_Goals
```

```python
# Penalty Goal

Total_Penalty_Goals = epl_data['Penalty_Goals'].sum()
Total_Penalty_Goals
```

```python
# Penalty Attempt

Total_Penalty_Attempts = epl_data['Penalty_Attempted'].sum()
Total_Penalty_Attempts
```

```python
# Pie Chart for Penalty Scored & Missed

plt.figure(figsize=(13, 6))
panelty_not_scored = epl_data['Penalty_Attempted'].sum() - Total_Penalty_Goals
data = [panelty_not_scored, Total_Penalty_Goals]
labels = ['Penalty Missed', 'Penalty Scored']
color = sns.color_palette('pastel')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show
```

```python
# Unique Position

epl_data['Position'].unique()
```

```python
# Forward players
    
epl_data[epl_data['Position'] == 'FW']    
```

```python
# Player From Different Nations

np.size((epl_data['Nationality'].unique()))
```

```python
# Most Players from which country

nationality = epl_data.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize = (12,6), color = sns.color_palette("magma"))
```

```python
# Clubs with Maximum Players in their Squad

epl_data['Club'].value_counts().nlargest(5).plot(kind = "bar", color=sns.color_palette("viridis"))
```

```python
# Clubs with least Players in their Squad

epl_data['Club'].value_counts().nsmallest(5).plot(kind = "bar", color=sns.color_palette("viridis"))
```

```python
# Players Based on Age Group
Under20 = epl_data[epl_data['Age'] <= 20] 
age20_25 = epl_data[(epl_data['Age'] > 20) & (epl_data['Age'] <= 25)] 
age25_30 = epl_data[(epl_data['Age'] > 25) & (epl_data['Age'] <= 30)] 
Above30 = epl_data[epl_data['Age'] > 30]
```

```python
x = np.array([Under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),Above30['Name'].count()])
mylabels = ["<=20", ">20 & <=25", "<25 & <=30", ">30"]
plt.title('Total Players with Age Groups',fontsize = 20)
plt.pie(x, labels = mylabels, autopct = '%.1f%%')
plt.show()
```

```python
# Under 20 Players in each Club

Players_under_20 = epl_data[epl_data['Age'] < 20]
Players_under_20['Club'].value_counts().plot(kind = 'bar', color = sns.color_palette("cubehelix"))
```

```python
# Players Under 20 in ManU

Players_under_20[Players_under_20["Club"] == 'Manchester United']
```

```python
# Average Age of Players in each Club

plt.figure(figsize=(12, 6))
sns.boxplot(x = 'Club', y = 'Age', data = epl_data)
plt.xticks(rotation = 90)
```

```python
num_player = epl_data.groupby('Club').size()
data = (epl_data.groupby('Club')['Age'].sum()) / num_player
data.sort_values(ascending = False)
```

```python
# Top 10 assist

Top_10_assists = epl_data[['Name', 'Club', 'Assists', 'Matches']].nlargest(n=10, columns = 'Assists')
Top_10_assists
```

```python
# Club Goals

Goals_by_clubs = pd.DataFrame(epl_data.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data=Goals_by_clubs.sort_values(by="Goals"),palette='rocket')
ax.set_xlabel("Club",fontsize=40)
ax.set_ylabel("Goals",fontsize=25)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs vs Total Goals',fontsize = 20)                 
```

```python
# Most Goals by Players

Top_goal_scorers = epl_data[['Name', 'Club', 'Goals', 'Matches']].nlargest(n=10, columns='Goals')
Top_goal_scorers
```

```python
Top_goal_scorers_per_match = epl_data[['Name', 'GoalsPerMatch', 'Matches', 'Goals']].nlargest(n=10, columns='GoalsPerMatch')
Top_goal_scorers_per_match
```

```python
# Most number of Matches played

Matches_played = epl_data[['Name', 'Club', 'Assists', 'Matches']].nlargest(n=15, columns = 'Matches')
Matches_played
```

```python
# Most Yellow card

epl_yellow = epl_data.sort_values(by='Yellow_Cards', ascending=False)[:10]
plt.figure(figsize=(20, 6))
plt.title("Players with the most yellow cards")
c = sns.barplot(x=epl_yellow['Name'], y = epl_yellow['Yellow_Cards'], label='Players', color='yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c
```
